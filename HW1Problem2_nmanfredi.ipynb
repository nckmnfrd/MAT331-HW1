```python
fig = plt.figure()
ax = plt.axes(projection='3d')



zaxe = np.linspace(0,20,1000)
xaxe = np.sin(zaxe)
yaxe = np.cos(zaxe)
ax.plot3D(xaxe, yaxe, zaxe, 'Red')

zaxe2 = np.linspace(0,20,1000)
xaxe2 = np.cos(zaxe2)
yaxe2 = np.sin(zaxe2)
ax.plot3D(xaxe2, yaxe2, zaxe2, 'Blue')
```
